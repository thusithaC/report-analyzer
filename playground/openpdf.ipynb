```python
from config import logger, REPORTS_FILE_LOCATION
file_loc = "/home/thusitha/work/bigdata/annual_reports/docs/LOLC.N0000/Annual_Report_2018-2019.pdf"

from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer
```

```python
pages = {}
for i, page_layout in enumerate(extract_pages(file_loc)):
    text_elements = []
    for element in page_layout:
        if isinstance(element, LTTextContainer):
            text_elements.append(element.get_text())
    pages[i] = text_elements
```

```python
# filepaths
import os

all_dirs = os.listdir(REPORTS_FILE_LOCATION)
all_files = []
for loc in all_dirs:
    files = [x for x in os.listdir(os.path.join(REPORTS_FILE_LOCATION, loc)) if str(x).endswith(".pdf")]
    all_files.extend(files)
```

```python
import re
# res = re.search(r'.*(20\d\d).?(20\d\d).*|.*(20\d\d).?(\d\d).*|.*(20\d\d).*', text)

errors = []
year_pattens = []
for file_name in all_files:
    print(file_name)
    res = re.search(r'.*(20\d\d).?(20\d\d).*|.*(20\d\d).?(\d\d).*|.*(20\d\d).*', file_name)
    year_patten = [x for x in res.groups() if x is not None]
    if len(year_patten) == 0:
        print(f"{file_name} cannot be matched")
        errors.append(file_name)
    else:
        print(year_patten)
        year_pattens.append(year_patten)
```
